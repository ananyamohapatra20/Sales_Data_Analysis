```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv("Diwali Sales Data.csv",encoding= 'unicode_escape')
```

```python
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df["Amount"]=df["Amount"].astype(int)
```

**<h1>Explanatory Data Analysis<h1>**

<h2>Gender</h2
>

```python
#Gender Count
ax=sns.countplot(x="Gender",data=df)
for i in ax.containers:
    ax.bar_label(i)
```

```python
#Gender Vs Amount Spent
sales=df.groupby(['Gender'],as_index=False)['Amount'].sum()
sns.barplot(x="Gender",y="Amount",data=sales)
```

*From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men*


<h2>Age<h2>


```python
ax=sns.countplot(data=df,x='Age Group',hue='Gender')
for i in ax.containers:
    ax.bar_label(i)
```

```python
sales_age=df.groupby(['Age Group'], as_index=False)['Amount'].sum()
sns.barplot(x='Age Group',y='Amount',data=sales_age)
```

From the analysis we can see that the people in the age group 26-35 are purchasing more.

<h2>State</h2>


```python
#Orders from different states
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by="Orders",ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,6)})
sns.barplot(data=sales_state,x='State',y='Orders')
```

```python
#Total Amount of Sales from different States
sales_amt_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by="Amount",ascending=False).head(10)
sns.barplot(data=sales_amt_state,x="State",y="Amount")
```

From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively.

<h2>Marital Status</h2>

```python
ax= sns.countplot(data=df,x="Marital_Status",hue='Gender')
for i in ax.containers:
    ax.bar_label(i)
```

```python
sales_mar=df.groupby(['Marital_Status','Gender'],as_index=False)["Amount"].sum()
sns.barplot(data=sales_mar,x='Marital_Status',y="Amount",hue="Gender")
```

From the above analysis it is seen that unmarried women have more purchasing Power.

<h2>Occupation</h2>

```python
ax=sns.countplot(data=df,x="Occupation")
sns.set(rc={'figure.figsize':(25,7)})
for i in ax.containers:
    ax.bar_label(i)
```

```python
sales_occ=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(data=sales_occ,x="Occupation",y="Amount")
```

Major people who are purchasing are from IT Sector,HealthCare,Aviation etc.

<h2>Product Category</h2>

```python
ax=sns.countplot(data=df,x='Product_Category')
for i in ax.containers:
    ax.bar_label(i)
```

```python
sales_prod=df.groupby(["Product_Category"],as_index=False)["Amount"].sum().sort_values(by='Amount', ascending=False).head(10)
sns.barplot(data=sales_prod,x="Product_Category",y="Amount")
```

From the above graph it is seen that the top selling products are food,clothing & Apparel,Electronics & Gadgets etc..

<h1>Conclusion</h1>

*Married women age group 26-35 yrs from UP,  Maharastra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing and Electronics category*
